```python
import math
import turtle
# nhap du lieu
xa = int(input('Nhap toa do xa: '))
ya = int(input('Nhap toa do ya: '))
xb = int(input('Nhap toa do xb: '))
yb = int(input('Nhap toa do yb: '))
xc = int(input('Nhap toa do xc: '))
yc = int(input('Nhap toa do yc: '))

# do dai cac canh tam giac
canhA = BC = math.sqrt((xc-xb)**2 + (yc-yb)**2)
canhB = AC = math.sqrt((xa-xc)**2 + (ya-yc)**2)
canhC = BA = math.sqrt((xb-xa)**2 + (yb-ya)**2)

# kiem tra cac canh de xac dinh tam giac hay khong
if (canhA + canhB > canhC and canhA + canhC > canhB and canhB + canhC > canhA):
    print("Đây là tam giác")
    
    # vẽ tam giác
    t = turtle.Turtle()
    t.shape('turtle')
    t.pensize(1)
    t.pencolor('red')
    t.fillcolor('pink')
    t.color('blue')
    t.penup()
    
    t.goto(xa,ya)
    t.pendown()
    t.begin_fill()
    t.circle(3)
    t.end_fill()
    t.goto(xa,ya+5)
    t.write('A', font=('Arial', 12))
    t.goto(xa,ya)
    t.goto(xb,yb)
    t.begin_fill()
    t.circle(3)
    t.end_fill()
    t.goto(xb,yb+5)
    t.write('B', font=('Arial', 12))
    t.goto(xb,yb)
    t.goto(xc,yc)
    t.begin_fill()
    t.circle(3)
    t.end_fill()
    t.goto(xc,yc+5)
    t.write('C', font=('Arial', 12))
    t.goto(xc,yc)
    t.goto(xa,ya)
    
    t.penup()
    
    # chu vi tam giac
    p = (canhA + canhB + canhC) / 2
    
    # duong cao từ các đỉnh
    hA = 2 * ((math.sqrt(p*(p-canhA)*(p-canhB)*(p-canhC))) / canhA)
    hB = 2 * ((math.sqrt(p*(p-canhA)*(p-canhB)*(p-canhC))) / canhB)
    hC = 2 * ((math.sqrt(p*(p-canhA)*(p-canhB)*(p-canhC))) / canhC)
    
    # tọa độ chân đường cao điểm AA 
    # (xAA - xa)*(xc - xb) + (yAA - ya)*(yc-yb) = 0 
    # 
    
    
    # 
    
    
    # kết thúc vẽ
    t.hideturtle()
    turtle.done()
    
else:
    print("Đây không phải là tam giác")

```
